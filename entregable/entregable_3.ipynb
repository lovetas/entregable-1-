```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

df = pd.read_csv('Tratados_internacionales_de_Colombia_20241206.csv')

lugares_adopcion= df['Lugar de Adopcion']

frecuencia_lugares= lugares_adopcion.value_counts().nlargest(10)

plt.figure(figsize=(9, 9))
plt.pie(frecuencia_lugares.values, labels=frecuencia_lugares.index, autopct='%1.1f%%')
plt.title('Lugares de adopción de tratados internacionales')
plt.axis('equal')  


plt.show()
```

```python


# Crear gráfica de pastel
plt.figure(figsize=(12, 11),
 plt.pie
    frecuencia_lugares.values,
    labels=frecuencia_lugares.index,
    autopct='%1.1f%%',
    startangle=140,
    
    wedgeprops={'edgecolor': 'black'},
    explode=[0.1 if i == 0 else 0 for i in range(len(frecuencia_lugares))]
)
plt.title('Lugares de adopción de tratados internacionales', fontsize=16, fontweight='bold')
plt.axis('equal')  # Asegura que sea un círculo perfecto
plt.legend(title="Lugares", loc="best", bbox_to_anchor=(1, 0.5))
plt.tight_layout()  # Ajusta el espacio para que no se sobrepongan elementos
plt.show()
```
