```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df.columns
```

```python
lugares_adopcion= df['Lugar de Adopcion']
```

```python
frecuencia_lugares= lugares_adopcion.value_counts().nlargest(10)
```

```python


plt.figure(figsize=(9, 9))
plt.pie(frecuencia_lugares.values, labels=frecuencia_lugares.index, autopct='%1.1f%%')
plt.title('Lugares de adopción de tratados internacionales')
plt.axis('equal')  


plt.show()
```

```python
#de esta forma de grafica de pastel, podemos saber cuales fueron la mayoria de los mas nombrados etre los primeros
#10 mas nombrados de la tabla.

#(usando matiplotlib)
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df.columns
```

```python
naturaleza = df['Naturaleza del Tratado']
```

```python
plt.figure(figsize=(11, 8))
naturaleza.value_counts().plot(kind='bar', color ='green')
plt.title('Distribución de tratados internacionales por naturaleza')
plt.xlabel('Naturaleza')
plt.ylabel('Cantidad de tratados')
plt.xticks(rotation=90)

plt.show()
```

```python
#de esta forma podemos ver la diferencia en donde se hizo el tratado y
# donde hubo un acuerdo en los tratados internacionales

# (usando seaborn)
```
