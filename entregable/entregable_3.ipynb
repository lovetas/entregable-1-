```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

df = pd.read_csv('Tratados_internacionales_de_Colombia_20241206.csv')

lugares_adopcion= df['Lugar de Adopcion']

frecuencia_lugares= lugares_adopcion.value_counts().nlargest(10)

plt.figure(figsize=(9, 9))
plt.pie(frecuencia_lugares.values, labels=frecuencia_lugares.index, autopct='%1.1f%%')
plt.title('Lugares de adopcin de tratados internacionales')
plt.axis('equal')  


plt.show()
```

```python
"""
preguntas a sacar que se podrian responder
-cual es porcentaje y decimal de cada uno?
-cuales fueron los 10 mas nombrados ?

-HIPOTESIS

Con esta informacion podriamos ver los paisaes o ciudades mas dominantes en los tratados internacionales y ver la cantidad de participaciones
que puede tener cada uno medido en procentajes.
"""
```

```python

plt.figure(figsize=(12, 11))
plt.pie(
    frecuencia_lugares.values,
    labels=frecuencia_lugares.index,
    autopct='%1.1f%%',
    startangle=140,
    
    explode=[0.1 if i == 0 else 0 for i in range(len(frecuencia_lugares))]
)
plt.title('Lugares mas nombrados de tratados internacionales', fontsize=16, fontweight='bold' )
plt.axis('equal')  # Asegura que sea un crculo perfecto
plt.legend(title="Lugares", loc="best", bbox_to_anchor=(1, 0.5))

plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

naturaleza = df['Naturaleza del Tratado']

plt.figure(figsize=(11, 8))
naturaleza.value_counts().plot(kind='bar', color ='green')
plt.title('Distribucin de tratados internacionales por naturaleza')
plt.xlabel('Naturaleza')
plt.ylabel('Cantidad de tratados')
plt.xticks(rotation=90)

plt.show()
```

```python
naturaleza = df['Naturaleza del Tratado']

plt.figure(figsize=(11, 8))
naturaleza.value_counts().plot(kind='bar', color ='aqua')
plt.title('Distribucin de tratados internacionales por naturaleza', fontsize=18, fontweight='bold')
plt.xlabel('Naturaleza')
plt.ylabel('Cantidad de tratados')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```
